```python
import pandas as pd
data = pd.read_csv('aqi.csv')
```

```python
df = pd.DataFrame(data)
df.drop_duplicates(inplace=True)
columns = df.columns.tolist()
print(f"Columns in the dataset: {columns}")
print(df["state"].unique())
```

```python
for sate in df['state'].unique():
    state_data = df[df['state'] == sate]
    avg_aqi = state_data['aqi_value'].mean()
    round_aqi = round(avg_aqi , 2)
    print(f"Average AQI for {sate}: {round_aqi}")
```
